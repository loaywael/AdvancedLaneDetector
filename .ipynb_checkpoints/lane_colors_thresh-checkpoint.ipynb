# Color Thresholding

```python
import cv2
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = plt.imread("./xyGrads.png")
fig = plt.figure(figsize=(15, 9))
hlsImg = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
h, l, s = cv2.split(hlsImg)

axis = fig.add_subplot(2, 2, 1)
axis.imshow(img)
axis = fig.add_subplot(2, 2, 2)
axis.imshow(h, "gray")
axis = fig.add_subplot(2, 2, 3)
axis.imshow(l, "gray")
axis = fig.add_subplot(2, 2, 4)
axis.imshow(s, "gray")
```

```python
def getSatThreshMask(hlsImg, minThresh, maxThresh=255):
    sChannel = hlsImg[:, :, 2]
    satMask = np.zeros_like(sChannel)
    satMask[(sChannel >= minThresh) & (sChannel <= maxThresh)] = 1
    return satMask


def getHueThreshMask(hlsImg, minThresh, maxThresh=255):
    hChannel = hlsImg[:, :, 0]
    hueMask = np.zeros_like(hChannel)
    hueMask[(hChannel <= minThresh) & (hChannel >= 0)] = 1
    return hueMask


def getRGThreshMask(rgbImg, minThresh, maxThresh=255):
    r, g = rgbImg[:, :, 0], rgbImg[:, :, 1]
    rgMask = np.zeros_like(r) 
    rgMask[((r >= minThresh) & (r <= maxThresh)) & ((g >= 0.75*minThresh) & (g <= maxThresh))] = 1
    return rgMask
```

```python
img = plt.imread("./edge_thresh_calib/frame1037.jpg")
hlsImg = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

fig = plt.figure(figsize=(15, 9))
satMask = getSatThreshMask(hlsImg, 130)
rgMask = getRGThreshMask(img, 175)
hueMask = getHueThreshMask(hlsImg, 30)
hsMask = cv2.bitwise_and(hueMask, satMask)

axis = fig.add_subplot(1, 2, 1)
axis.imshow(hsMask, "gray")
axis = fig.add_subplot(1, 2, 2)
axis.imshow(rgMask, "gray")
```

```python
fig = plt.figure(figsize=(15, 15))
imgsPaths = glob("./edge_thresh_calib/*.jpg")
for i, imgPath in enumerate(imgsPaths):
    img = plt.imread(imgPath)
    hlsImg = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    satMask = getSatThreshMask(hlsImg, 125)
    rgMask = getRGThreshMask(img, 200)
    hueMask = getHueThreshMask(hlsImg, 30)
    hsMask = cv2.bitwise_and(hueMask, satMask)
    lanesMask = np.zeros_like(rgMask
    lanesMask[(hsMask == 1) | (rgMask == 1)] = 1
    axis = fig.add_subplot(len(imgsPaths)//2, 2, i+1)
    axis.imshow(lanesMask, "gray")
```

# Twice time efficient color thresholding

```python
fig = plt.figure(figsize=(15, 15))
imgsPaths = glob("./edge_thresh_calib/*.jpg")
for i, imgPath in enumerate(imgsPaths):
    img = plt.imread(imgPath)
    hlsImg = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    
    _, satMask = cv2.threshold(hlsImg[:, :, 2], 175, 255, cv2.THRESH_BINARY)
    _, hueMask = cv2.threshold(hlsImg[:, :, 2], 33, 255, cv2.THRESH_BINARY_INV)
    _, redMask = cv2.threshold(img[:, :, 0], 170, 255, cv2.THRESH_BINARY)
    _, greenMask = cv2.threshold(img[:, :, 1], 170, 255, cv2.THRESH_BINARY)
    hueSatMask = cv2.bitwise_and(hueMask, satMask)
    redGreenMask = cv2.bitwise_and(redMask, greenMask)
    laneMask = cv2.bitwise_or(redGreenMask, hueSatMask)
#     rgMask = getRGThreshMask(img, 200)
#     hueMask = getHueThreshMask(hlsImg, 30)
#     lanesMask = np.zeros_like(rgMask)
#     lanesMask[(hsMask == 1) | (rgMask == 1)] = 1
    axis = fig.add_subplot(len(imgsPaths)//2, 2, i+1)
    axis.imshow(laneMask, "gray")
```
